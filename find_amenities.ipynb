```python
import pandas as pd
import json
from tqdm import tqdm
import math

from return_nearby_amenities import search_nearby_amenities, valid_amenities
```

```python
listings = pd.read_csv('./property_guru/search/processed/20231116/processed.csv')
listings.head()
```

```python
for amenity in tqdm(valid_amenities):
    listings[f'has_{amenity}'] = False
    listings[f'{amenity}'] = [[]]*len(listings)
    for i, row in tqdm(listings.iterrows()):
        if math.isnan(row['latitude']):
            continue
        
        result = search_nearby_amenities(
            lat1=row['latitude'],
            long1=row['longitude'],
            type=amenity
        )

        listings.at[i, f'has_{amenity}'] = (len(result)>0)

        listings.at[i, f'{amenity}'] = result
```

```python
listings.to_csv('./final.csv')
```
